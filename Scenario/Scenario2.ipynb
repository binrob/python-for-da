```python
import pandas as pd
import numpy as np
survey = pd.DataFrame({
    'response_id': [1, 2, 3, 4, 5, 6, 7, 8, 3, 10, 5],
    'customer_id': ['C001', 'C002', 'C003', None, 'C005', 'C006', 
                    'C007', 'C008', 'C003', 'C010', 'C005'],
    'age': [25, 30, np.nan, 28, 35, np.nan, 42, 29, 35, np.nan, 35],
    'satisfaction': [5, 4, 5, np.nan, 3, 4, 5, np.nan, 5, 3, 3],
    'recommend': ['Yes', 'Yes', 'Yes', 'No', np.nan, 'Yes', 
                  'Yes', 'Yes', 'Yes', np.nan, 'Maybe'],
    'comment': ['Great!', None, 'Excellent', None, 'Ok', None, 
                'Love it', None, 'Excellent', 'Not bad', 'Ok'],
    'submit_date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04',
                    '2024-01-05', '2024-01-06', '2024-01-07', '2024-01-08',
                    '2024-01-09', '2024-01-10', '2024-01-11']
})
```

```python
#task 1 
survey.duplicated().sum() # 0 response duplicated
survey.duplicated(subset=['customer_id']).sum() # 2 customers have mutiple responses
missing_data_column_percentage = (survey.isnull().sum()/ len(survey)) *100
print(missing_data_column_percentage)
```

```python
#task 2 
cleaned = survey.copy()
cleaned = cleaned.drop_duplicates(subset=['response_id'], keep='last')
cleaned = cleaned.dropna(subset=['customer_id', 'satisfaction'])
median_age = cleaned['age'].median()
cleaned['age'] = (cleaned['age'].fillna(median_age))
cleaned['recommend'] = cleaned['recommend'].fillna('Unknown')
print(cleaned)
```

```python
#task3 
remained = cleaned['response_id'].count() # -> 7
avg_satis_score = cleaned['satisfaction'].mean() #-> ~4.143/5
percentage_recommend = ( (cleaned['recommend'] == 'Yes').sum() / len(cleaned['recommend'])) *100 #-> 71.43%
def category_age (age):
    if age >40:
        return 'over 40'
    elif age <30 :
        return 'under 30'
    else:
        return '30-40'
cleaned['age_group'] = cleaned['age'].apply(category_age)
cleaned.groupby('age_group')['satisfaction'].mean().nlargest()
```
