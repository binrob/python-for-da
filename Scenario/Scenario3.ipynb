```python
import pandas as pd
import numpy as np

employees = pd.DataFrame({
    'employee_id': ['E001', 'E002', 'E003', 'E004', 'E002', 'E005', 
                    'E006', 'E003', 'E007', None, 'E008', 'E004'],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Bob', 'Eve',
             'Frank', 'Charlie', 'Grace', 'Henry', 'Ivy', 'David'],
    'department': ['Sales', 'IT', 'Sales', None, 'IT', 'HR',
                   'Sales', 'Sales', 'IT', 'Sales', None, 'Marketing'],
    'salary': [50000, 60000, 55000, np.nan, 62000, 48000,
               53000, 55000, np.nan, 51000, 47000, 58000],
    'performance_score': [8.5, 7.2, np.nan, 8.0, 7.5, 9.0,
                         7.8, 9.2, 8.3, np.nan, 8.7, 8.0],
    'years_experience': [3, 5, 4, np.nan, 5, 2, 3, 4, 6, 3, np.nan, 7],
    'last_review_date': ['2024-01-15', '2024-01-10', '2024-01-12', '2024-01-08',
                        '2024-02-15', '2024-01-20', '2024-01-18', '2024-02-10',
                        '2024-01-25', '2024-01-22', '2024-01-28', '2024-02-05']
})
```

```python
#task 2
#step 1 : handle duplicate
cleaned = employees.copy()
cleaned = cleaned.drop_duplicates(subset='employee_id', keep= 'last')
cleaned = cleaned.dropna(subset=['employee_id','performance_score'])
#step 2 : missing data
cleaned['department'] = cleaned['department'].fillna('Unassigned')
dept_medians = cleaned.groupby('department')['salary'].median()
cleaned['salary'] = cleaned.apply(lambda  row: dept_medians[row['department']] if pd.isna(row['salary']) else row['salary'], axis=1)
# i take your hint to do that sentence above though i dont clearly know the meaning yet
cleaned['years_experience'] = cleaned['years_experience'].fillna(0)
```

```python
#task 3
avg_salary_per_department = cleaned.groupby('department')['salary'].mean()
print(avg_salary_per_department.idxmax())
# Is there correlation between years_experience and performance_score? (qualitative observation) idk this one
three_top_performers=cleaned['performance_score'].sort_values(ascending=False).head(3)
```
