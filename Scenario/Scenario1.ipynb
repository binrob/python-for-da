```python
import pandas as pd
import numpy as np

orders = pd.DataFrame({
    'order_id': [1001, 1002, 1003, 1004, 1002, 1005, 1006, 1003, 1007, 1008],
    'customer_email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', 
                       'david@email.com', 'bob@email.com', None, 
                       'frank@email.com', 'charlie@email.com', 'grace@email.com', None],
    'product': ['Laptop', 'Mouse', None, 'Keyboard', 'Mouse', 
                'Monitor', 'Laptop', 'Keyboard', None, 'Headphones'],
    'price': [1200, 25, 75, np.nan, 25, 300, 1200, 75, 150, np.nan],
    'quantity': [1, 2, np.nan, 1, 2, 1, 1, np.nan, 2, 1],
    'order_date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', 
                   '2024-01-02', '2024-01-05', '2024-01-06', '2024-01-03', 
                   '2024-01-07', '2024-01-08'],
    'status': ['completed', 'completed', 'pending', 'completed', 
               'completed', 'pending', 'completed', 'pending', 'completed', 'pending']
})
```

```python
#task 1
orders.duplicated().sum()
orders.duplicated(subset=['order_id']).sum()
#Which columns have missing data and what percentage?
(orders.isnull().sum() / len(orders)) *100
```

```python
#task 2 
task2 = orders.copy()
task2=task2.drop_duplicates(subset=['order_id'], keep='first')
task2['customer_email']= task2['customer_email'].fillna('guest@unknown.com')
task2=task2.dropna(subset = ['product','price'])
task2['quantity']=task2['quantity'].fillna(1)
task2
```

```python
#task 3
q1 = task2['order_id'].count()
q2 = task2.duplicated(subset=['order_id']).sum()
print(q2)
q3 = task2[['product','price']].isnull().sum()
task2['revenue'] = task2['price'] * task2['quantity']
print(task2) 
```

```python
#task 4
orders.groupby('customer_email')['order_id'].count().nlargest()
avg_order_value = task2['revenue'].mean()
print(avg_order_value)
```
