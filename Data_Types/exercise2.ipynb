```python
import pandas as pd
import numpy as np

# Messy employee data
employees = pd.DataFrame({
    'employee_id': ['E001', 'E002', 'E003', 'E004', 'E005', 'E006', 'E007', 'E008'],
    'name': ['  ALICE johnson  ', 'bob  SMITH', '  Charlie BROWN  ', 'DAVID wilson',
             'eve  DAVIS', '  FRANK miller  ', 'grace MOORE', '  HENRY taylor  '],
    'email': ['Alice.Johnson@COMPANY.com', 'BOB.smith@company.COM', 'charlie.brown@COMPANY.com',
              'DAVID.WILSON@company.com', 'Eve.Davis@COMPANY.COM', 'frank.miller@company.com',
              'Grace.Moore@COMPANY.com', 'henry.TAYLOR@company.COM'],
    'department': ['  sales  ', 'IT', '  SALES  ', 'it', 'HR  ', 'sales', '  IT  ', 'hr'],
    'salary': [50000, 60000, 55000, 62000, 48000, 52000, 250000, 51000]
})
```

```python
#task 1 
employees['name'] = employees['name'].str.strip().str.title()
employees['email'] = employees['email'].str.lower()
employees['department'] = employees['department'].str.strip().str.upper()
```

```python
#task 2 
Q1 = employees['salary'].quantile(0.25)
Q3 = employees['salary'].quantile(0.75)
IQR = Q3 - Q1
Lower_bound = Q1 - 1.5 * IQR
Upper_bound = Q3 + 1.5 * IQR
outliners = employees[(employees['salary'] < Lower_bound) | (employees['salary'] > Upper_bound)]
print(Q1, Q3, IQR, Lower_bound, Upper_bound, "\n" )
print("Outliers based on salary:\n" , outliners)
```

```python
#task 3
cleaned_employees = employees[(employees['salary'] >= Lower_bound) & (employees['salary'] <= Upper_bound)]
employees['amount_capped'] = employees['salary'].clip(lower=Lower_bound, upper=Upper_bound)
employees['is_outlier'] = np.where((employees['salary'] < Lower_bound) | (employees['salary'] > Upper_bound), True, False)
```

```python
#task 4
mean_salary = employees['salary'].mean()
median_salary = employees['salary'].median()
num_employees = employees.shape[0]
print(num_employees)
#i think Cap is the best handling method because the outlier can be a director or CEO with a high salary
```
