```python
import pandas as pd
import numpy as np
from datetime import datetime
raw_orders = pd.DataFrame({
    'order_id': ['1001', '1002', '1003', '1004', '1005'],
    'order_date': ['2024-01-15', '2024-02-20', 'Invalid', '2024-03-10', '2024-04-05'],
    'customer_id': ['C001', 'C002', 'C003', 'C004', 'C005'],
    'product_price': ['1200.50', '25', 'N/A', '75.99', '300'],
    'quantity': ['2', '5', '3', '1', '2'],
    'discount_percent': ['10', '5', '0', '15', '20'],
    'is_member': ['Yes', 'No', 'Yes', 'Yes', 'No']
})

```

```python
#task1
raw_orders['order_id'] = pd.to_numeric(raw_orders['order_id'])
raw_orders['order_date'] = pd.to_datetime(raw_orders['order_date'], errors= 'coerce')
raw_orders['product_price'] = pd.to_numeric(raw_orders['product_price'] , errors='coerce')
raw_orders['product_price'] = raw_orders['product_price'].astype(float)
raw_orders['quantity'] = pd.to_numeric(raw_orders['quantity'])
raw_orders['discount_percent'] = raw_orders['discount_percent'].astype(float)
raw_orders['is_member'] = raw_orders['is_member'].map({'Yes': True , 'No': False})
print(raw_orders)
```

```python
#task2 
raw_orders=raw_orders.dropna(subset=['order_date']) 
raw_orders= raw_orders.fillna(raw_orders['product_price'].median())
raw_orders['final_price']= raw_orders['product_price'] * raw_orders['quantity'] * (1 - (raw_orders['discount_percent']/100))
print(raw_orders)
```

```python
#task 3 
raw_orders['order_month'] = raw_orders['order_date'].dt.month # type: ignore
raw_orders['order_year'] = raw_orders['order_date'].dt.year # type: ignore
now = datetime.now()
raw_orders['days_since_order'] = (now - raw_orders['order_date']).dt.days # type: ignore
print(raw_orders)
print(raw_orders.dtypes)
```
