```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [25, 30, np.nan, 35],
    'salary': [50000, 60000, 55000, np.nan],
    'department': ['Sales', 'IT', 'Sales', 'IT']
})
df['Senior'] = df['age'].apply(lambda x: 'True' if x > 28 else 'False')
df['age'] = df['age'].apply(lambda row: 30 if pd.isna(row) else row)
df['total_comp'] = df.apply(lambda x : x['salary'] + (x['age'] *1000) , axis=1)
#df['category'] = df['salary'].apply(lambda x: 'High' if x > 55000 else 'Low' if x < 50000 else 'Medium')
def categorize_salary (x):
    if x > 55000 :
        return 'High'
    elif x < 50000:
        return 'Low'
    elif pd.isna(x):
        return 'Unknown'
    else : 
        return 'Medium'
df['category'] = df['salary'].apply(categorize_salary)
df
```
