```python
import pandas as pd
import numpy as np
df = pd.read_excel("Excel_Practice_Project#1.xlsx")
```

```python
df['Customer_Name']= df['Customer_Name'].str.strip()
df['Unit_Price']= df['Unit_Price'].replace('[\$,]', '', regex=True).astype(float) #ignore dollar sign and convert to float
df['Discount']= df['Discount'].replace('[\%,]', '', regex=True).astype(float) #ignore percentage sign and convert to float
df['Order_Date']= pd.to_datetime(df['Order_Date'], format='%m/%d/%Y') #convert to datetime format
df.to_excel("Cleaned_Excel_Practice_Project#1.xlsx", index=False)
```
